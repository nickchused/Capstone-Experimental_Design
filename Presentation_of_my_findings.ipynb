```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
```

```python
PATH = 'https://datahub.io/core/global-temp/r/annual.csv'
temp_data = pd.read_csv(PATH)
temp_data.info()
temp_data.head()
```

The following evaluations show how close the distributions of temperatures for each time period follow a normal distribution.  As a general rule, if the skewness variable is between -1 and 1, and the kurtosis level is between -2 and 2, the distributions can be treated and analyzed as normal distributions.

```python
print(stats.describe(temp_1880_1919['Mean']))
print(stats.describe(temp_1920_1959['Mean']))
print(stats.describe(temp_1960_1999['Mean']))
print(stats.describe(temp_2000_2016['Mean']))
```

Based on skewness and kurtosis the distributions can be treated and analyzed as normal distributions.

The Shapiro-Wilk test is another method of testing if distributions are normally distributed.  The Shaprio-Wilk test gives two values.  If the first value, W, is close to 1, and the second value, the p-value, is greater than 0.05, we can treat and analyze the distributions as normal distributions.  

```python
print(stats.shapiro(temp_1880_1919['Mean']))
print(stats.shapiro(temp_1920_1959['Mean']))
print(stats.shapiro(temp_1960_1999['Mean']))
print(stats.shapiro(temp_2000_2016['Mean']))
```

Based on ther results of the Shapiro-Wilk test, the distributions can be treated and ananlyzed as normal distributions.

The collection of boxplots below shows the averge temperature anomolies for each year from 1880 to 2016 as compared to corresponing 1951 to 1980 temperature means.

Based on this collection of boxplots, it appears there is an undeniable upward trend in temperature anomalies over the last 137 years.
